```python
# Load the data.

# readMhd from utils.py from the LNDb Challenge
# Reference: https://lndb.grand-challenge.org/
def readMhd(filename):
    # read mhd/raw image
    itkimage = sitk.ReadImage(filename)
    scan = sitk.GetArrayFromImage(itkimage) #3D image
    spacing = itkimage.GetSpacing() #voxelsize
    origin = itkimage.GetOrigin() #world coordinates of origin
    transfmat = itkimage.GetDirection() #3D rotation matrix
    return scan,spacing,origin,transfmat
```

```python
# assume we are in the same directory as the data

# Get the filenames for the images and the labels into arrays
imgFilenames = np.array(list(filter(lambda filename: filename.split('.')[-1] == 'mhd' and 'segmentation' not in filename.split('.')[0], os.listdir())))
```

```python
lblFilenames = np.array(list(filter(lambda filename: filename.split('.')[-1] == 'mhd' and 'segmentation' in filename.split('.')[0], os.listdir())))
```

```python
# Load the data using the readMhd method
```

```python
scans = []
spacings = []
origins = []
transfmats = []

for filename in imgFilenames:
    scan, spacing, origin, transfmat = readMhd(filename)
    scans.append(scan)
    spacings.append(spacing)
    origins.append(origin)
    transfmats.append(transfmat)

scans = np.array(scans)
spacings = np.array(spacings)
origins = np.array(origins)
transfmats = np.array(transfmats)
```

```python
labels = []
lblSpacings = []
lblOrigins = []
lblTransfmats = []

for filename in lblFilenames:
    scan, spacing, origin, transfmat = readMhd(filename)
    labels.append(scan)
    lblSpacings.append(spacing)
    lblOrigins.append(origin)
    lblTransfmats.append(transfmat)

labels = np.array(labels)
lblSpacings = np.array(lblSpacings)
lblOrigins = np.array(lblOrigins)
lblTransfmats = np.array(lblTransfmats)
```

```python
# Part d., Prostate volume
# image and label spacing are the same for each scan
```

```python
# Count the active (non-zero) voxels
activeVoxels = [np.count_nonzero(label) for label in labels]
activeVoxels = np.array(activeVoxels)
```

```python
# Calculate the average spacing for each scan
avgSpacings = [spacing.mean() for spacing in spacings]
avgSpacings = np.array(avgSpacings)
```

```python
# Estimate the volume in cubic centimeters
estVol = activeVoxels * avgSpacings * 0.001
```

```python
plt.plot(estVol)
plt.title('Part d: Estimated Volume Among Cases')
plt.xlabel('Case Number')
plt.ylabel('Volume (cubic cm)')
```

```python
refSpacings = []
refAxis = []
for spacing in spacings:
    x, y, z = spacing
    #print(x, y, z)
    if np.round(x, 3) == np.round(y, 3):
        refSpacings.append(x)
        refAxis.append('x')
    else:
        refSpacings.append(z)
        refAxis.append('z')
```

```python
df = pd.DataFrame(np.array([range(50), estVol, avgSpacings, refSpacings, refAxis]).T, columns=['Case Number','Esimated Volume (cubic cm)', 'Average Spacing', 'Reference Spacing', 'Reference Axis'])
```
